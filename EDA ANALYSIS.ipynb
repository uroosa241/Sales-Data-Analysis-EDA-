```python
Q1 = df['amount'].quantile(0.25)
Q3 = df['amount'].quantile(0.75)

IQR = Q3 - Q1

lower_limit = Q1 - 1.5 * IQR
upper_limit = Q3 + 1.5 * IQR

# Remove outliers
df_no_outliers = df[
    (df['amount'] >= lower_limit) &
    (df['amount'] <= upper_limit)
]
```

```python
sns.boxplot(
    y=df['amount'],
    flierprops={
        'marker': 'o',
        'markerfacecolor': 'red',
        'markeredgecolor': 'black',
        'markersize': 10
    }
)

plt.show()
```

```python
sns.histplot(df['amount'], bins=30, kde=True)
plt.title("sales distribution")
plt.show()
```

```python
sns.barplot(x=df["gender"], y=df["amount"])
plt.title("Total sales by Gender")
plt.show()
```

Female customers generated higher total sales compared to male customers.
Marketing campaigns can be targeted toward the higher-spending gender segment.

```python
sns.barplot(x='age-group', y='amount', data=df, estimator=sum)
plt.xticks(rotation=45)
plt.title("Sales by Age Group")
plt.show()
```

Customers in the highest bar age group contribute the most revenue.
This age segment represents the most valuable customer base.

```python
df['age_bin'] = pd.cut(df['age'], bins=[0,18,25,35,45,60,100])

plt.figure(figsize=(10,5))
sns.barplot(x='age_bin', y='amount', data=df, estimator='sum')
plt.xticks(rotation=45)
plt.title("Age VS spending")
plt.show()
```

A few product categories dominate overall revenue generation.
Inventory and promotions should focus on these high-performing categories.

```python
sns.barplot(x="marital_status", y="orders", data=df)
plt.title("Marital status vs orders")
plt.show()
```

```python
sns.barplot(x="state", y="amount", data=df.head(10), estimator=sum)
plt.title("State vs amount")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10,5))
df.groupby("product_category")["amount"].sum().sort_values(ascending=False).plot(kind="bar")
plt.title("Product category vs amount")
plt.xticks(rotation=45)
plt.show()
```

One marital group shows higher purchasing activity than the other.
Customer preferences may vary based on marital status.

```python
plt.figure(figsize=(10,5))
df.groupby("zone")["orders"].sum().sort_values(ascending=False).plot(kind="bar")
plt.title("Zone vs orders")
plt.xticks(rotation=45)
plt.show()
```

The top-performing zone generates the largest share of revenue.
Sales strategies from this zone can be replicated in lower-performing zones.

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))

corr_matrix = df.corr(numeric_only=True)

sns.heatmap(
    corr_matrix,
    annot=True,      # show correlation values
    fmt=".2f",       # 2 decimal places
    cmap="coolwarm",
    linewidths=0.5,
    square=True
)

plt.title("Correlation Matrix", fontsize=16)
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)

plt.tight_layout()
plt.show()
```

The correlation analysis shows very weak relationships among the numerical variables. Age, marital status, and number of orders have negligible correlation with sales amount. Therefore, sales performance appears to be driven more by categorical factors such as state, gender, occupation, and product category rather than the available numerical variables.
